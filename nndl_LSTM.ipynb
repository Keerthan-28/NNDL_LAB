```python
import numpy as np
# Assuming sigmoid and tanh functions are available from previous cells (anlb12lcW9-0)
# If running this cell independently, ensure they are defined or imported.

class LSTMcell:
    def __init__(self, input_size, hidden_size):
        self.input_size = input_size
        self.hidden_size = hidden_size

        # weights for gates
        self.Wf = np.random.randn(hidden_size, hidden_size + input_size)  # forget gate
        self.Wi = np.random.randn(hidden_size, hidden_size + input_size)  # input gate
        self.Wc = np.random.randn(hidden_size, hidden_size + input_size)  # cell state
        self.Wo = np.random.randn(hidden_size, hidden_size + input_size)  # output gate

        # biases
        self.bf = np.zeros((hidden_size, 1))
        self.bi = np.zeros((hidden_size, 1))
        self.bc = np.zeros((hidden_size, 1))
        self.bo = np.zeros((hidden_size, 1))

    def forward(self, x, h_prev, c_prev):
        # concatenate h_prev & input x
        combined = np.vstack((h_prev, x))

        # 1. forget gate
        f_t = sigmoid(np.dot(self.Wf, combined) + self.bf)

        # 2. input gate
        i_t = sigmoid(np.dot(self.Wi, combined) + self.bi)
        C_s = tanh(np.dot(self.Wc, combined) + self.bc)

        # 3. cell state update
        C_t = f_t * c_prev + i_t * C_s

        # 4. output gate
        o_t = sigmoid(np.dot(self.Wo, combined) + self.bo)
        h_t = o_t * tanh(C_t)

        return h_t, C_t, f_t, i_t, o_t


np.random.seed(1)

input_size=1
hidden_size=3
sequence_length=4

#sample input sequence
X= [np.random.randn(input_size,1) for _ in range(sequence_length)]

#initializeLSTM
lstm=LSTMcell(input_size,hidden_size)

#initial states
h= np.zeros((hidden_size,1))
C=np.zeros((hidden_size,1))

#forward pass thorugh sequence
print("step by step LSTM EXECUTION:\n")

for t in range(sequence_length):
    x= X[t]
    h,C,f,i,o= lstm.forward(x,h,C)

    print(f"Time step{t+1}")
    print("Input:", x.flatten())
    print("Forget gate:", f.flatten())
    print("Input gate:", i.flatten())
    print("Output gate:", o.flatten())
    print("hidden state:", h.flatten())
    print("-"*40)
```
